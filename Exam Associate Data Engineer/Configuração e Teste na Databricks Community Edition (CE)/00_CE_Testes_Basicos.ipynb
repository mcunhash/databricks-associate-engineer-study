## Crie um DataFrame simples com PySpark

```python
data = [("Alice", 1), ("Bob", 2), ("Charlie", 3)]
df = spark.createDataFrame(data, ["Name", "ID"])
print("Schema do DataFrame:")
df.printSchema()
print("Conteúdo do DataFrame:")
df.display() # Use display() para uma visualização bonita no Databricks
```

## Crie uma view temporária e consulte-a

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW pessoas_ce AS
SELECT 'David' AS Nome, 4 AS Idade
UNION ALL
SELECT 'Eva' AS Nome, 5 AS Idade;

SELECT * FROM pessoas_ce;
```
