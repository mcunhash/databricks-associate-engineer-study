# # Módulo 1: Plataforma de Inteligência Databricks

## 1.1. Arquitetura Lakehouse

*   **Definição:** A Lakehouse Architecture combina os melhores aspectos dos Data Lakes (flexibilidade, baixo custo, suporte a dados não estruturados) e dos Data Warehouses (estrutura, transações ACID, governança, performance).
*   **Por que é importante?** Elimina silos de dados, simplifica a arquitetura, permite que dados de diferentes tipos sejam usados para BI e Machine Learning na mesma plataforma.

## 1.2. Componentes Fundamentais

### 1.2.1. Delta Lake
*   **O que é:** Um formato de armazenamento de código aberto que estende os arquivos Parquet com um log de transações.
*   **Benefícios:**
    *   **ACID Transactions:** Garante consistência e isolamento de dados.
    *   **Schema Enforcement & Evolution:** Ajuda a manter a qualidade dos dados e a adaptar o schema ao longo do tempo.
    *   **Time Travel:** Permite acessar versões históricas dos dados.
    *   **Upserts (MERGE INTO):** Facilita a atualização e inserção de dados de forma eficiente.

### 1.2.2. Unity Catalog
*   **O que é:** Uma solução unificada de governança para dados e IA no Databricks.
*   **Benefícios:**
    *   **Metadados Centralizados:** Um único lugar para gerenciar metadados de tabelas, views, volumes.
    *   **Segurança Granular:** Controle de acesso baseado em SQL para tabelas, colunas, linhas.
    *   **Linhagem de Dados:** Rastreia como os dados são transformados.
    *   **Descoberta de Dados:** Facilita encontrar e entender os dados disponíveis.

### 1.2.3. Databricks Runtime (DBR)
*   **O que é:** Otimizações e bibliotecas pré-instaladas que rodam em cima do Apache Spark.
*   **Benefícios:** Melhor performance, maior confiabilidade, bibliotecas atualizadas (ex: Delta Lake, MLflow).

## 1.3. Tipos de Compute no Databricks

*   **All-Purpose Clusters (Interactive):**
    *   **Uso:** Desenvolvimento interativo, exploração de dados, notebooks.
    *   **Características:** Podem ser compartilhados, permanecem ativos por um tempo.
    *   **(Seu cluster na CE é um exemplo disso)**
*   **Job Clusters:**
    *   **Uso:** Execução automatizada de jobs (pipelines ETL/ELT, treinamentos de ML).
    *   **Características:** Criados sob demanda para o job, terminam após a execução, otimizados para custo/performance de jobs.
    *   **(Você os usará ao criar Jobs no ambiente corporativo)**
*   **SQL Warehouses:**
    *   **Uso:** Cargas de trabalho SQL, BI, dashboards.
    *   **Características:** Otimizados para consultas SQL, auto-scaling rápido, compatíveis com ferramentas de BI.
    *   **(Você tem um no ambiente corporativo)**

## 1.4. Visão Geral do Workspace
*   **Explore o menu lateral esquerdo:**
