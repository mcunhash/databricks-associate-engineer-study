#  Módulo 4: Produzindo Pipelines de Dados

## 1. Introdução aos Databricks Jobs

### 1.1. O que são Databricks Jobs?
 **Databricks Jobs** são a principal forma de automatizar e agendar a execução de tarefas (como notebooks, scripts Python ou JARs) no Databricks. Eles transformam o código que você desenvolve interativamente em um processo de produção robusto, confiável e automatizado.

**Por que usar Jobs?**
-   **Automação:** Executam tarefas repetitivas sem intervenção manual.

-   **Confiabilidade:** Gerenciam a execução, retries e monitoramento para garantir o sucesso.

-   **Orquestração:** Permitem definir a ordem e dependências entre múltiplas etapas do seu pipeline.

-   **Monitoramento:** Oferecem logs detalhados e alertas para você acompanhar o status.

### 1.2. Componentes Principais de um Databricks Job
Um Databricks Job é composto por vários elementos-chave que definem seu comportamento e funcionalidade:

- **Tarefas (Tasks):** São as unidades de trabalho que o Job executará (e.g., um Notebook Databricks, um script Python, um JAR).

-  **Cluster:** O ambiente de computação onde as tarefas serão executadas. Para Jobs, geralmente se usa um "Job Cluster" que é provisionado e terminado automaticamente para otimizar custos e isolamento.

-   **Agendamento (Schedule):** Define quando o Job deve ser executado (e.g., diário, semanal, por evento).

-   **Dependências (Dependencies):** Em Jobs com múltiplas tarefas, você pode definir a ordem de execução.

-   **Parâmetros (Parameters):** Permite passar valores dinâmicos para as tarefas do Job.

-   **Notificações (Notifications):** Configuração de alertas para sucesso, falha ou timeouts.
